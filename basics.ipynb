```python
data=pd.read_csv("C:\\Users\\Abhinav Raj\\Downloads\\countries.csv")
```

```python
data
```

```python
# compare the population growth in US and China
```

```python
import numpy as np 
np.unique(data.country)
```

```python
US=data[data.country=='United States']
```

```python
US
```

```python
China=data[data.country=="China"]
```

```python
China
```

```python
plt.plot(US.year,US.population/ 10**6)
plt.plot(China.year,China.population/ 10**6)
plt.title("US population over time")
plt.legend(["US","China"])
plt.xlabel("Year")
plt.ylabel("Population in millions")
plt.show()
```

```python
# Relative population growth
```

```python
US.population/US.population.iloc[0] *100
```

```python
China.population/China.population.iloc[0] *100
```

```python
plt.plot(US.year,US.population/US.population.iloc[0] *100)
plt.plot(China.year,China.population/China.population.iloc[0] *100)
plt.title("US population over time")
plt.legend(["US","China"])
plt.xlabel("Year")
plt.ylabel("Population growth (first year =100)")
plt.show()
```
