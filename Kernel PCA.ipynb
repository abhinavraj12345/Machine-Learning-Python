```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=50,random_state=1)
plt.scatter(X[y==0,0],X[y==0,1],color='red',marker='o',alpha=0.5)
plt.scatter(X[y==1,0],X[y==1,1],color='blue',marker='o',alpha=0.5)
plt.ylabel('y cordinate')
plt.xlabel('x cordinate')
plt.show()
```

## Simple PCA

```python
from mlxtend.feature_extraction import PrincipalComponentAnalysis as PCA
pca=PCA(n_components=2)
X_pca=pca.fit(X).transform(X)
```

```python
plt.scatter(X_pca[y==0,0],X_pca[y==0,1],color='red',marker='o',alpha=0.5)
plt.scatter(X_pca[y==1,0],X_pca[y==1,1],color='blue',marker='o',alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## Kernel PCA

```python
from mlxtend.feature_extraction import RBFKernelPCA as KPCA
kpca=KPCA(gamma=15.0,n_components=2)
kpca.fit(X)
X_kpca=kpca.X_projected_
```

```python
plt.scatter(X_kpca[y==0,0],X_kpca[y==0,1],color='red',marker='o')
plt.scatter(X_kpca[y==1,0],X_kpca[y==1,1],color='blue',marker='o')
plt.xlabel('Principal component 1')
plt.ylabel('Principal Component 2')
plt.title('Kernel Principal Component Analysis using RBF Kernel Function')
```
